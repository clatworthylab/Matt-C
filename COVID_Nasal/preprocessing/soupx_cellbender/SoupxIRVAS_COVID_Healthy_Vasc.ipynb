```r
## program ##
soupx <- function(dataDirs) {
	kelvinny::dirCreate(paste0('./soupx/', dataDirs))
	system(paste0('cp ', dataDirs, '/filtered_feature_bc_matrix/barcodes.tsv.gz ./soupx/', dataDirs, '/barcodes.tsv.gz'))
	system(paste0('cp ', dataDirs, '/filtered_feature_bc_matrix/features.tsv.gz ./soupx/', dataDirs, '/features.tsv.gz'))
	require(SoupX)
	require(Matrix)
	require(Seurat)
	dat <- Read10X(paste0(dataDirs, '/filtered_feature_bc_matrix/'))
	seu <- tryCatch(CreateSeuratObject(counts = dat), error = function(e) {CreateSeuratObject(counts = dat$`Gene Expression`)})
	seu[["percent_mt"]] <- PercentageFeatureSet(seu, pattern = "^MT-")
	seu <- NormalizeData(seu)
	seu <- FindVariableFeatures(seu)
	seu <- ScaleData(seu, vars.to.regress = c("percent_mt", 'nCount_RNA'))
	seu <- RunPCA(seu, features = VariableFeatures(object = seu))
	seu <- FindNeighbors(seu)
	seu <- FindClusters(seu, algorithm = 4)
	clusters <- Idents(seu)
	sc <- load10X(dataDirs)	
	sc <- setClusters(sc, clusters)
	sc = autoEstCont(sc, doPlot = FALSE)
	out = adjustCounts(sc, roundToInt=TRUE)
	writeMM(out, paste0("./soupx/", dataDirs, "/matrix.mtx"))
	system(paste0('gzip ', './soupx/', dataDirs, '/matrix.mtx'))
}
```
